```python
#Criando numeros aleatorios e msotrando a dimensao da matriz
b= np.random.rand(5,5)
print(b)

#print da dimensao
print('dimensao da matriz')
print(b.shape)
```

```python
#Criando Matriz de valores aleatorios entre uma faixa estabelecida
vmin = 1
vmax = 100
c= vmin+(vmax-vmin)*b
print(c)
```

```python
#Plotar matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha 

x=np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:])
plt.title('Grafico com Valores Inteiros no Eixo X')
plt.xticks(x)
```

```python
#Plotando apenas a primeira linha
x=np.array(range(0,np.size(c[0,:])))
print(x)
plt.scatter(x,c[0,:])
```

```python
# Criando um graf de barras e exportando

fig, ax=plt.subplots(3,1)  #plota ax pq Ã© um objeto plt
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# exportar
import os
rootFolder = os.getcwd()
print(rootFolder) #Ate aqui criou a imagem mas importa pra nenhuma pasta do meu notebook
```
